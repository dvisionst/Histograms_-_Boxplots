# Histograms & Boxplots Exercise (Core)


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
data = '/content/catsvdogs.xlsx'
cat_n_dogs_df = pd.read_excel(data)
```

## 1. Plot a histogram of the Percentage of Dog Owners.

```python
plt.style.use('seaborn')
cat_n_dogs_df['Percentage of Dog Owners'].hist(bins=7, edgecolor='black')
plt.xlabel('Percentage of Animal Owners')
plt.ylabel('Count');
```

## 2. Plot a histogram of the Percentage of Cat Owners.

```python
plt.style.use('seaborn')
cat_n_dogs_df['Percentage of Cat Owners'].hist(bins=15, edgecolor='black')
plt.xlabel('Percentage of Animal Owners')
plt.ylabel('Count');
```

## 3a. What can we see by comparing these two histograms? What information does this tell us?

The percentage of dog owners is unimodal and Left Skewed. This means that the mode is closer to the right and greater than the median or mean. The percentage of cat owners is more bell shaped unimodal. 

```python
plt.style.use('seaborn')
cat_n_dogs_df['Percentage of Dog Owners'].hist(bins=7, edgecolor='black', alpha=0.5, legend=True)
cat_n_dogs_df['Percentage of Cat Owners'].hist(bins=15, edgecolor='black', alpha=0.5, legend=True)
plt.xlabel('Percentage of Animal Owners')
plt.ylabel('Count')
plt.legend(['Dogs', 'Cats']);
```

## 4. Create two boxplots on the same plot: one of the mean number of cats per household and the other of the mean number of dogs. An example of what this might look like is given below:



```python

dogs = cat_n_dogs_df['Mean Number of Dogs per household']
cats = cat_n_dogs_df['Mean Number of Cats']
boxplots = plt.boxplot([cats, dogs],
                        notch=True,
                        labels=['Cats', 'Dogs'],
                        patch_artist=True,
                        widths=.6,
                        medianprops= dict(linestyle='-', linewidth=2, color='Yellow'),
                        boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .5)
                        );

boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('orange')

plt.ylabel('Mean Number of Animals per State', fontsize = 20);

plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);


```

## 5. What can we see by comparing these two boxplots? What information does this tell us?

It shows that the median of cats per state is higher than that of dogs. 
It also shows tha the min value of cats is higher than the median of dogs.
Overall this suggests that their is a higher percentage of cats owners  than there are dog owners.
